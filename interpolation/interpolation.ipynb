# UPSAMPLING


```python
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt
```

1 - Low-res image creation

```python
img_HR = cv2.imread('cameraman.png', cv2.IMREAD_GRAYSCALE)
print('Original Dimensions : ',img_HR.shape)

scale_percent = 25 # percent of original size
width_HR = int(img_HR.shape[1])
height_HR = int(img_HR.shape[0])
dim_HR = (width_HR, height_HR)


scale_percent = 25 # percent of original size
width_LR = int(img_HR.shape[1] * scale_percent / 100)
height_LR = int(img_HR.shape[0] * scale_percent / 100)
dim_LR = (width_LR, height_LR)

img_LR = cv2.pyrDown(img_HR, dim_LR)
print('Original Dimensions : ',img_LR.shape)


plt.figure()
plt.imshow(img_HR, cmap = "gray")
plt.axis('off')
plt.title('HR image')

plt.figure()
plt.imshow(img_LR, cmap = "gray")
plt.axis('off')
plt.title('LR image')
```

2 - Interpolation

```python
# resize image
img_interpolated_NN = cv2.resize(img_LR, dim_HR, interpolation = cv2.INTER_NEAREST)
img_interpolated_bilinear = cv2.resize(img_LR, dim_HR, interpolation = cv2.INTER_LINEAR)
img_interpolated_area = cv2.resize(img_LR, dim_HR, interpolation = cv2.INTER_AREA)
img_interpolated_bicubic = cv2.resize(img_LR, dim_HR, interpolation = cv2.INTER_CUBIC)
img_interpolated_Lanczos = cv2.resize(img_LR, dim_HR, interpolation = cv2.INTER_LANCZOS4) 

img_interpolated_pyrUp = img_LR = cv2.pyrUp(img_LR, dstsize=dim_HR)
```

```python
# Original image
plt.figure()
plt.imshow(img_HR, cmap = "gray")
plt.axis('off')
plt.title('Nearest Neighbors based upscaled image')

# Upscaled image using Nearest Neighbors
plt.figure()
plt.imshow(img_interpolated_NN, cmap = "gray")
plt.axis('off')
plt.title('Nearest Neighbors based upscaled image')

# Upscaled image using Bilinear Interpolation
plt.figure()
plt.imshow(img_interpolated_bilinear, cmap = "gray")
plt.axis('off')
plt.title('Bilinear Interpolation based upscaled image')

# Upscaled image using Area Relation
plt.figure()
plt.imshow(img_interpolated_area, cmap = "gray")
plt.axis('off')
plt.title('Area Relation based upscaled image')

# Upscaled image using Bicubic Interpolation
plt.figure()
plt.imshow(img_interpolated_bicubic, cmap = "gray")
plt.axis('off')
plt.title('Bicubic Interpolation based upscaled image')

# Upscaled image using 	Lanczos interpolation
plt.figure()
plt.imshow(img_interpolated_Lanczos, cmap = "gray")
plt.axis('off')
plt.title('Lanczos interpolation based upscaled image')

# Upscaled image using 	Lanczos interpolation
plt.figure()
plt.imshow(img_interpolated_pyrUp, cmap = "gray")
plt.axis('off')
plt.title('Gaussian Pyramid based upscaled image')
```
